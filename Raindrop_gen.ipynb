```python
width = 1280
height = 1024

image_list=os.listdir('/content/images/1o')
image_list=list(map(lambda x:'/content/images/1o/'+x,image_list))

for i in trange(len(image_list)):
  filepath=image_list[i]
  filename=filepath.split('/')[-1]
  List_of_Drops, label_map  = generate_label(height, width, cfg)
  mask_path='/content/1o_mask/'+filename
  gen_path='/content/1o_gen_images/'+filename

  output_image, output_label, mask = generateDrops(filepath, cfg, List_of_Drops)

  mask.save(mask_path)
  output_image.save(gen_path)


```
